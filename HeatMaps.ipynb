```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Diabetes_combined_data.csv')


df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce')
df.drop(columns=['Date', 'Time'], inplace=True)

pivot_df = df.pivot_table(index='DateTime', columns='Code', values='Value', aggfunc='first')
pivot_df = pivot_df.apply(pd.to_numeric, errors='coerce')


corr_matrix = pivot_df.corr()

# Human-readable code labels
code_labels = {
    0: 'No Record',
    4: 'Unspecified Blood Glucose',
    33: 'Regular Insulin (U)',
    34: 'NPH Insulin (U)',
    35: 'Ultralente Insulin (U)',
    36: 'Lispro Insulin (U)',
    48: 'Ultralente Insulin (U)',
    56: 'Unspecified Insulin (U)',
    57: 'Unspecified Glucose (mg/dL)',
    58: 'Blood Glucose (mg/dL)',
    59: 'Finger Stick Glucose (mg/dL)',
    60: 'Dietary Intake (kcal)',
    61: 'Unspecified Meal',
    62: 'Exercise (min)',
    63: 'Stress Event',
    64: 'Illness Event',
    65: 'Hypoglycemia Symptoms',
    66: 'Other Symptoms',
    67: 'Ketone Presence',
    68: 'Alcohol Use',
    69: 'High Blood Sugar Symptoms',
    70: 'Insulin Reaction',
    71: 'Unusual Activity',
    72: 'Menstruation'
}

corr_matrix.rename(index=code_labels, columns=code_labels, inplace=True)


plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap (Human-Readable Labels)", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")

correlation_matrix = df.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap (BRFSS 2015 Health Indicators)", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
